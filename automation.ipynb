```python
import pandas as pd

print("===== Superstore Sales Automation Started =====")


# Step 1: Load Raw Dataset

df = pd.read_csv("Superstore.csv")
print("Dataset loaded successfully.")


# Step 2: Data Cleaning


# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Fill missing values
df.ffill(inplace=True)

# Convert Order Date to datetime
df["Order Date"] = pd.to_datetime(df["Order Date"])

print("Data cleaning completed.")


# Step 3: Save Cleaned Dataset

df.to_csv("Processed_Superstore.csv", index=False)

print("Processed_Superstore.csv created successfully.")

# Step 4: Calculate KPIs

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_quantity = df["Quantity"].sum()
total_orders = df["Order ID"].nunique()
average_discount = df["Discount"].mean()

kpi = pd.DataFrame({
    "KPI": [
        "Total Sales",
        "Total Profit",
        "Total Quantity",
        "Total Orders",
        "Average Discount"
    ],
    "Value": [
        total_sales,
        total_profit,
        total_quantity,
        total_orders,
        average_discount
    ]
})

print("\n===== Key Performance Indicators =====")
print(kpi)

with pd.ExcelWriter("Processed_Superstore.xlsx", engine="openpyxl") as writer:
    df.to_excel(writer, sheet_name="Cleaned Data", index=False)
    kpi.to_excel(writer, sheet_name="KPIs", index=False)

print("Processed_Superstore.xlsx created successfully.")

print("\n===== Automation Completed Successfully =====")
```
